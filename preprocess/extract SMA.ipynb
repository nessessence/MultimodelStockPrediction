```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
os.listdir('../data')
```

```python
df = pd.read_csv('../data/dowjone1985_2019.csv')
```

```python
windows = [5, 10, 25, 75, 125, 200]
for window in windows:
    df['sma_' + str(window)] = df['Close'].rolling(window=window).mean()
```

```python
cols = ['Close']
for window in windows:
    cols.append('sma_' + str(window))
df.plot(y=cols)
```
